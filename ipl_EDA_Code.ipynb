```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Load the dataset
df = pd.read_csv('/content/Book_ipl22_ver_33.csv')

```

```python
# Check the first 5 records
print(df.head())
```

```python

# Check the shape of the dataset
print("Shape of the dataset:", df.shape)
```

```python

# Check the summary statistics of the dataset
print("Summary Statistics: \n", df.describe())
```

```python

# Check for missing values
print("Missing Values: \n", df.isnull().sum())
```

```python

# Plot a bar graph of match winners
plt.bar(df['match_winner'].value_counts().index, df['match_winner'].value_counts().values)
plt.xlabel("Match Winners", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Match Winners Count", fontsize=16)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python

# Plot a histogram of first innings score
plt.hist(df['first_ings_score'], bins = 30)
plt.xlabel("First Innings Score")
plt.ylabel("Count")
plt.title("First Innings Score Histogram")
plt.show()
```

```python

# Plot a histogram of player of the match count
plt.hist(df['player_of_the_match'].value_counts().values, bins = 30)
plt.xlabel("Player of the match count")
plt.ylabel("Count")
plt.title("Player of the match count Histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt

# data for wins and losses
teams = ['Chennai', 'Kolkata', 'Delhi', 'Banglore', 'Gujarat', 'Hyderabad', 'Rajasthan', 'Punjab']
wins = [4, 5, 5, 4, 5, 4, 4, 4]
losses = [3, 2, 1, 2, 1, 2, 2, 3]

# plotting the pie chart
fig, ax = plt.subplots()
ax.pie(wins, labels=teams, autopct='%1.1f%%', shadow=True, startangle=90)
ax.set_title("Teams with Most Wins")
# plotting the pie chart for losses
fig, ax = plt.subplots()
ax.pie(losses, labels=teams, autopct='%1.1f%%', shadow=True, startangle=90)
ax.set_title("Teams with Most Losses")

# displaying the pie chart
plt.show()

```
